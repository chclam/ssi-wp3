```python
import pandas as pd
import os

ssi_data_directory = "/nfs/workspace/data/ssi"
ssi_data_directory
```

```python
final_preprocessing_dir = os.path.join(ssi_data_directory, "preprocessing/05-final")
os.listdir(final_preprocessing_dir)
```

```python
data_files = [os.path.join(final_preprocessing_dir, filename) for filename in os.listdir(final_preprocessing_dir)]
data_files
```

```python
lidl_df = pd.read_parquet(data_files[2])
lidl_df.head()
```

```python
plus_df = pd.read_parquet(data_files[0])
plus_df.head()
```

```python
ah_df = pd.read_parquet(data_files[1])
ah_df.head()
```

```python
lidl_receipts = lidl_df.receipt_text.unique()
plus_receipts = plus_df.receipt_text.unique()
ah_receipts = ah_df.receipt_text.unique()
```

```python
lidl_receipts.shape, plus_receipts.shape, ah_receipts.shape
```

```python
from transformers import AutoTokenizer

old_tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```python
def get_unique_tokens(tokenizer, receipt_texts):
    receipt_tokens = [tokenizer.tokenize(receipt_text)
    for receipt_text in receipt_texts]
   
    return {token for token_list in receipt_tokens for token in token_list}    
```

```python
lidl_unique_tokens = get_unique_tokens(old_tokenizer, lidl_receipts)
plus_unique_tokens = get_unique_tokens(old_tokenizer, plus_receipts)
ah_unique_tokens = get_unique_tokens(old_tokenizer, ah_receipts)
```

```python
def overlap(left_set, right_set):
    return len(left_set.intersection(right_set)) / (len(left_set) + len(right_set))

overlap_lidl_plus = overlap(lidl_unique_tokens, plus_unique_tokens)
overlap_lidl_ah = overlap(lidl_unique_tokens, ah_unique_tokens)
overlap_ah_plus = overlap(ah_unique_tokens, plus_unique_tokens)

overlap_lidl_plus, overlap_lidl_ah, overlap_ah_plus
```

# Train our own tokenizer

```python
import numpy as np

all_receipts = np.concatenate([lidl_receipts, plus_receipts, ah_receipts])
all_receipts[:-10]
```

```python
tokenizer = old_tokenizer.train_new_from_iterator(all_receipts, 10000)
```

```python
own_lidl_unique_tokens = get_unique_tokens(tokenizer, lidl_receipts)
own_plus_unique_tokens = get_unique_tokens(tokenizer, plus_receipts)
own_ah_unique_tokens = get_unique_tokens(tokenizer, ah_receipts)
```

```python
own_overlap_lidl_plus = overlap(own_lidl_unique_tokens, own_plus_unique_tokens)
own_overlap_lidl_ah = overlap(own_lidl_unique_tokens, own_ah_unique_tokens)
own_overlap_ah_plus = overlap(own_ah_unique_tokens, own_plus_unique_tokens)

own_overlap_lidl_plus, own_overlap_lidl_ah, own_overlap_ah_plus
```
